```python
# As categorical data cannot be used to learn our model and to plot the heatmap, 
# so I have used categorical encoding to encode each of
# the categorical types taken by each categorical feature for plotting heatmap because
# with one hot encoding the heat map becomes very large and ugly looking 
# and used one hot encoding for learning the model.
```

```python
df['workclass'] = df['workclass'].astype('category')
df['workclass'] = df['workclass'].cat.codes
df['education'] = df['education'].astype('category')
df['education'] = df['education'].cat.codes
df['occupation'] = df['occupation'].astype('category')
df['occupation'] = df['occupation'].cat.codes
df['relationship'] = df['relationship'].astype('category')
df['relationship'] = df['relationship'].cat.codes
df['race'] = df['race'].astype('category')
df['race'] = df['race'].cat.codes
df['sex'] = df['sex'].astype('category')
df['sex'] = df['sex'].cat.codes
df['native.country'] = df['native.country'].astype('category')
df['native.country'] = df['native.country'].cat.codes
df['income'] = df['income'].astype('category')
df['income'] = df['income'].cat.codes
```

```python
dfnew = pd.read_csv('adult.csv') # Loading the dataset into the program again for one hot encoding.
dfnew.head()
```

```python
#Dropping redundant features
```

```python
dfnew = dfnew.drop(columns=['marital.status'])
dfnew = dfnew.drop(columns=['education.num'])
dfnew.head()
```

```python
#Splitting the dataset into feature dataset and target dataset
```

```python
X = dfnew.drop(columns=['income'])
Y = dfnew.drop(columns=[ 'workclass'            
,'education', 'occupation', 'relationship', 'race', 'sex', 'native.country','age','fnlwgt','capital.gain','capital.loss','hours.per.week'])
X.head()
```

```python
Y.head()
```

```python
#Doing OneHotEncoding on the Independent variables
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
feature_array1 = ohe.fit_transform(X[['relationship']]).toarray()
feature_labels1 = ohe.categories_
feature_labels1 = np.array(feature_labels1).ravel()
features1 =pd.DataFrame(feature_array1, columns = feature_labels1)
X = X.drop(columns=['relationship'])
X1 = pd.concat([X,features1],axis=1)

feature_array2 = ohe.fit_transform(X1[['sex']]).toarray()
feature_labels2 = ohe.categories_
feature_labels2 = np.array(feature_labels2).ravel()
features2 =pd.DataFrame(feature_array2, columns = feature_labels2)
X1 = X1.drop(columns=['sex'])
X2 = pd.concat([X1,features2],axis=1)

feature_array3 = ohe.fit_transform(X2[['education']]).toarray()
feature_labels3 = ohe.categories_
feature_labels3 = np.array(feature_labels3).ravel()
features3 =pd.DataFrame(feature_array3, columns = feature_labels3)
X2 = X2.drop(columns=['education'])
X3 = pd.concat([X2,features3],axis=1)

feature_array4 = ohe.fit_transform(X3[['race']]).toarray()
feature_labels4 = ohe.categories_
feature_labels4 = np.array(feature_labels4).ravel()
features4 =pd.DataFrame(feature_array4, columns = feature_labels4)
X3 = X3.drop(columns=['race'])
X4 = pd.concat([X3,features4],axis=1)

feature_array5 = ohe.fit_transform(X4[['occupation']]).toarray()
feature_labels5 = ohe.categories_
feature_labels5 = np.array(feature_labels5).ravel()
features5 =pd.DataFrame(feature_array5, columns = feature_labels5)
X4 = X4.drop(columns=['occupation'])
X5 = pd.concat([X4,features5],axis=1)

feature_array6 = ohe.fit_transform(X5[['workclass']]).toarray()
feature_labels6 = ohe.categories_
feature_labels6 = np.array(feature_labels6).ravel()
features6 =pd.DataFrame(feature_array6, columns = feature_labels6)
X5 = X5.drop(columns=['workclass'])
X6 = pd.concat([X5,features6],axis=1)

feature_array7 = ohe.fit_transform(X6[['native.country']]).toarray()
feature_labels7 = ohe.categories_
feature_labels7 = np.array(feature_labels7).ravel()
features7 =pd.DataFrame(feature_array7, columns = feature_labels7)
X6 = X6.drop(columns=['native.country'])
X7 = pd.concat([X6,features7],axis=1)
```

```python
X7.head()
```

```python
#Doing OneHotEncoding on the target variable
```

```python
feature_array8 = ohe.fit_transform(Y[['income']]).toarray()
feature_labels8 = ohe.categories_
feature_labels8 = np.array(feature_labels8).ravel()
features8 =pd.DataFrame(feature_array8, columns = feature_labels8)
Y = Y.drop(columns=['income'])
Ynew = pd.concat([Y,features8],axis=1)
```

```python
Ynew.head()
```

```python
# Splitting the data into train, test and validation datasets.
from sklearn.model_selection import train_test_split
X_main, X_test, Y_main, Y_test = train_test_split(X7,Ynew,test_size=0.2,random_state =50) 
X_main1, X_val, Y_main1, Y_val = train_test_split(X_main,Y_main,test_size=0.2,random_state =50) 
```

```python
from sklearn.preprocessing import StandardScaler # employing standard normalization to normalize the variables.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_main1)
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)
```

```python
# Doing Hyperparameter experiments.
```

```python
# Creating Random Forest Classifier model.
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 110, max_features =3, max_depth=9, max_leaf_nodes=9 )
rf.fit(X_train_scaled,Y_main1)
```

```python
# Evaluating the model using mean accuracy score.
```

```python
rf.score(X_train_scaled,Y_main1)
```

```python
rf.score(X_test_scaled,Y_test)
```

```python
rf.score(X_val_scaled,Y_val)
```
